```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime

#menginport library yang di butuhkan di antranya adalah (pandas, matplotlib, numpy dan datatime)

now = datetime.datetime.now()
print('Hello, Waktu sekarang adalah {}'.format(now))
#untuk menampilkan waktu
```

```python
#untuk membaca file csv yang sudah di download

mydata = pd.read_csv('cars.csv', delimiter=';', nrows=20, skiprows=[1])
```

```python
#Descripsi kode mydata di atas...

# 1. nrwos 20 untuk menampilkan data sebanyak 20 rows dari file cars.csv
# 2. delimiter di sini digunakan utuk pemisah selain (;) bisa juga menggunakan (\t), dan titik dua (:).
#3. kiprows=[1] untuk menghilangkan rows tipe data dari data-data yang akan di tampilkan
```

```python
#menampilkan data yang ada pada file CSV dengan ketentuan di atas
mydata
```

```python
x = mydata['Weight'];
y = mydata['Horsepower']

#melakukan pendefenisian titk inteval dari data di atas di ambil titk Weight dan Horsepower
```

```python
plt.xlabel('Horsepower'); plt.ylabel('Weight')
plt.scatter(x,y)

#menampilkan parabola yang sudah di tentukan di atas
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
fig.suptitle('Contoh Beberapa Grafis Acak')
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

ax1.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
ax1.set_xlabel('Sumbu X')
ax1.set_ylabel('Sumbu Y')

x = np.linspace(0, 10, 30)
y = np.sin(x)
ax2.plot(x, y, '-ok')
ax2.set_xlabel('Sumbu X')
ax2.set_ylabel('Sumbu Y')

fig.tight_layout(rect=[0, 0.03, 1, 0.97])
plt.show()
```
